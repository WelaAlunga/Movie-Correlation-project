```python
#Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sb

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in data


df = pd.read_csv(r'C:\Users\WEL43591\OneDrive - Spectrum Health\Documents\Wela docs\LinkedIn Learning\Python\movies.csv')
```

```python
# Viewing the data


df.head()
```

```python
# Find missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Filter out missing values

df = df.dropna()
```

```python
# Check for missing values again
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# looking at Column Data types
df.dtypes
```

```python
# Changing column data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using searborn

sb.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# looking at correlation
```

```python
df.corr() #pearson by defualt
```

```python
# We can see a high correlation between budget and gross

correlation_matrix = df.corr()

sb.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# looking at Company
# Changing object variables to numeric 
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes

df_numeric.head()
```

```python
# Looking at correlation between all variables

correlation_matrix = df_numeric.corr()

sb.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numeric.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > .5]

high_corr
```
